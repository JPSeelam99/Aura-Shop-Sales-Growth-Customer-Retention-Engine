```python
# Check null counts column by column
print(customers.isnull().sum())
print(orders.isnull().sum())
print(order_items.isnull().sum())
```

```python
# Identification of missing values in order_delivered_customer_date column
```

```python
missing_count = orders['order_delivered_customer_date'].isnull().sum()
total_orders = len(orders)
print(f"Missing: {missing_count} out of {total_orders}")
```

```python
#Convert that count into a percentage
pct_missing = (missing_count / total_orders) * 100
print(f"{pct_missing:.2f}% of orders are missing a delivery date")
```

```python
# identifying the cause
missing_rows = orders[orders['order_delivered_customer_date'].isnull()]
print(missing_rows['order_status'].value_counts())
```
